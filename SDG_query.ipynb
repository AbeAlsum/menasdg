```python
import requests
import json
import pandas as pd
import csv, re
```

# TEST CSV

```python
url = 'https://unstats.un.org/SDGAPI/v1/sdg/Series/DataCSV'
headers = {"Content-Type": "application/x-www-form-urlencoded",
            "Accept": "application/octet-stream"}

myobj = {'seriesCodes': 'SI_POV_DAY1'}
x = requests.post(url, headers=headers, data=myobj)

print(x.text)
```

```python
for i in x.text:
    print(i)
```

```python
data = x.text.splitlines()

# Open the file for writing
with open("tmp.csv", "w") as csv_file:
    # Create the writer object with tab delimiter
    writer = csv.writer(csv_file, delimiter = '\t')
    for line in data:
        # Writerow() needs a list of data to be written, so split at all empty spaces in the line 
        writer.writerow(re.split('\s+',line))
```

```python
pd.read_csv('tmp.csv')
```

# GEO Area to list

```python
resp = requests.get('https://unstats.un.org/SDGAPI/v1/sdg/GeoArea/List').json()
```

```python
geoAreaCode = []
geoAreaName = []
for i in resp:
    geoAreaCode.append(i['geoAreaCode'])
    geoAreaName.append(i['geoAreaName'])

data = pd.DataFrame({"geoAreaCode" : geoAreaCode, 
                     "geoAreaName" : geoAreaName})
```

```python
data.to_excel('geoAreaName.xlsx', index = False)
```
